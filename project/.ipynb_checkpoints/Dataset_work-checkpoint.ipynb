```python
import pandas as pd
```

```python
df = pd.read_csv('Phishing_Email.csv')
```

```python
df.head()
```

```python
df['Email Type'] = df['Email Type'].replace({'Safe Email': 0, 'Phishing Email': 1})
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
print(df['Email Text'].sample(n=1).iloc[0])
```

```python
df['Email Text'] = df['Email Text'].str.lower()
```

```python
df['Email Text'] = df['Email Text'].str.strip()
```

```python
df['Email Text'] = df['Email Text'].str.replace(r'\s+', ' ', regex=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['Email Text']
y = df['Email Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=50000, min_df=2, max_df=0.8)
```

```python
X_train_tfidf = vectorizer.fit_transform(X_train)
```

```python
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
y_train_pred = model.predict(X_train_tfidf)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
print(f"Difference: {train_accuracy - accuracy:.4f}")
```

```python
test_email_clean = test_email.lower().strip()
test_email_tfidf = vectorizer.transform([test_email_clean])
prediction = model.predict(test_email_tfidf)
prediction_prob = model.predict_proba(test_email_tfidf)

print(f"Prediction: {prediction[0]}")
print(f"0 = Safe, 1 = Phishing")
print(f"\nConfidence:")
print(f"Safe: {prediction_prob[0][0]:.4f}")
print(f"Phishing: {prediction_prob[0][1]:.4f}")
```
